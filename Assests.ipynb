```python
crypto = crypto.reindex(trad.index)
alter = alter.reindex(trad.index)
```

```python
all_assets["Crypto"] = all_assets.Crypto.interpolate()
```

```python
returns = all_assets.pct_change().dropna()
returns
```

```python
plt.figure(figsize= (12, 8))
sns.violinplot(data = returns)
plt.show()
```

```python
import numpy as np
```

```python
ann_return = returns.mean() * 252
ann_return
```

```python
ann_risk = returns.std() * np.sqrt(252)
ann_risk
```

```python
plt.figure(figsize = (15, 12))
plt.scatter(x = ann_risk, y = ann_return)
for i in ann_risk.index:
    plt.annotate(i, xy=(ann_risk[i], ann_return[i]), size = 10)
plt.xlabel("annualized Risk")
plt.ylabel("annualized Return")
plt.show()
```

```python
corr_matrix = returns.corr()
corr_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, square = True, annot = True, cmap = "RdBu")
plt.show()
```
